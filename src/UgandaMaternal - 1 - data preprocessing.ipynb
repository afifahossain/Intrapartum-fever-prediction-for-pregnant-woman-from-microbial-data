```python
def create_full_data(abundance_data, sample_history):
  # from sklearn import preprocessing
  # label_encoder = preprocessing.LabelEncoder()
  # abundance_data['Unnamed: 0'] = label_encoder.fit_transform(abundance_data['Unnamed: 0'])
  # le_name_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

  tst = abundance_data.T
  tst.columns = tst.iloc[0,:]
  tst = tst.drop(['Unnamed: 0'])

  tst['Unnamed: 0'] = tst.index

  full_data = pd.merge(sample_history, tst, on='Unnamed: 0')

  return full_data
```

```python
data_1 = create_full_data(data_abundance_1, sample_details_1 )
print(data_1.shape)
data_1.head()
```

```python
data_2 = create_full_data(data_abundance_2, sample_details_2 ) 
print(data_2.shape)
data_2.head()
```

```python
data_2.groupby('Fever during labor')['Delivery Mode'].value_counts()
```

```python
full_data = pd.merge(data_1,data_2,how='outer',indicator=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np 

for c in data.columns[1:12]:
    #print "---- %s ---" % c
    if c == 'Fever during labor': 
      continue 
    #c = 'Age (years)'
    rs = pd.DataFrame(data.groupby('Fever during labor')[c].value_counts())
    #print(rs)
    # Convert the resulting series to a DataFrame
    df_counts = pd.DataFrame(rs)
    df_counts.columns = ['Count']
    # Reset the index to have 'Fever during labor' and 'c' as columns
    df_counts = df_counts.reset_index().sort_values(by=[c])
    #print(df_counts)


    # Create a pivot table to fill missing entries with 0 counts
    pivot_data = df_counts.pivot(index=c, columns='Fever during labor', values='Count').fillna(0)

    # Get the unique ages
    ages = pivot_data.index 

    # Set the width and positions for the bars
    bar_width = 0.2
    positions = np.arange(len(ages))

    # Plot the grouped bar plot
    plt.figure(figsize=(6, 4))

    for i, group in enumerate(pivot_data.columns):
        counts = pivot_data[group]
        plt.bar(positions + i * bar_width, counts, width=bar_width, label=group)

    plt.xlabel(c) 
    plt.ylabel('Count')
    plt.title(f'{c} Vs. Fever during labor')
    plt.xticks(positions + bar_width * (len(pivot_data.columns) - 1) / 2, ages, rotation=90)
    plt.legend()

    plt.show()

    
```
